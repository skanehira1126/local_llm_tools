```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embeddings = HuggingFaceEmbeddings(
    model_name="pkshatech/GLuCoSE-base-ja-v2",
    model_kwargs={"device": "mps"},
)
```

```python
embeddings.embed_documents([
    "query: Hello, my name is Alice",
])[:5]
```

```python
import chromadb
```

```python
client = chromadb.Client()
```

```python
collection = client.get_or_create_collection("my_collection")
```

```python
import markitdown
md = markitdown.MarkItDown()
document = md.convert("/Users/kanehirashinichi/Desktop/1511.05952v4.pdf").text_content
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=400,
    chunk_overlap=40,
)
```

```python
docs = [
    f"passage: {text}"
    for text in text_splitter.split_text(document)
]
```

```python
embed_docs = embeddings.embed_documents(docs)
```

```python
import uuid
```

```python
collection.add(
    documents=docs,
    embeddings=embed_docs,
    ids=[str(uuid.uuid4()) for _ in docs]
)
```

```python
query = "経験につける重みの考え方は？"
```

```python
embed_query = embeddings.embed_query(f"query: {query}")
```

```python
search_result = collection.query(
    query_embeddings=[embed_query],
    n_results=5,
)
```
